```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual


df = pd.read_excel("Neuro Journals.xlsx")
sd_dict = {'Impact Factor ': 0, \
            'H5': 0, \
            'H5 Median': 0, \
            'H-Index - 3': 0,\
            'G-Index - 3': 0, \
            'Cites per year - 3': 0, \
            'Cites per article - 3': 0, \
            'Avg. Cites of top 10 papers': 0, \
            'Avg. Altmetric Score of Top 10 Cited Papers': 0, \
            'Avg. Altmetric Score of Top 10 Cited Papers': 0, \
            '# of twitter followers': 0, \
            '# of tweets': 0}

mean_dict = {'Impact Factor ': 0, \
            'H5': 0, \
            'H5 Median': 0, \
            'H-Index - 3': 0,\
            'G-Index - 3': 0, \
            'Cites per year - 3': 0, \
            'Cites per article - 3': 0, \
            'Avg. Cites of top 10 papers': 0, \
            'Avg. Altmetric Score of Top 10 Cited Papers': 0, \
            'Avg. Altmetric Score of Top 10 Cited Papers': 0, \
            '# of twitter followers': 0, \
            '# of tweets': 0}

for col in sd_dict:

    sd_dict[col] =  df[col].std()
    mean_dict[col] = df[col].mean()


def select_jour(w1, w2=0, w3=0, w4=0, w5=0, w6=0, w7=0, w8=0, w9=0, w10=0, w11=0, w12=0):

  weights = {'Impact Factor ': w1, \
            'H5': w2, \
            'H5 Median': w3, \
            'H-Index - 3': w4,\
            'G-Index - 3': w5, \
            'Cites per year - 3': w6, \
            'Cites per article - 3': w7, \
            'Avg. Cites of top 10 papers': w8, \
            'Avg. Altmetric Score of Top 10 Cited Papers': w9, \
            'Avg. Altmetric Score of Top 10 Cited Papers': w10, \
            '# of twitter followers': w11, \
            '# of tweets': w12}


  global df
  for i in range(len(df)):

      score = 0
      for col in weights:

          if weights[col] != 0:

              attribute_val = df.loc[i, col]
              score += ((attribute_val - mean_dict[col]) / sd_dict[col]) * weights[col]

      df.loc[i, 'Calc_Score'] = score


  df = df.sort_values(by='Calc_Score', ascending=False)
  df = df.reset_index(drop=True)

  print(df[['JOURNAL NAME', 'Calc_Score']].head(20))


slider1 = widgets.IntSlider(description='Impact Factor', min=0, max=100, value=0)
slider2 = widgets.IntSlider(description='H5', min=0, max=100, value=0)
slider3 = widgets.IntSlider(description='H5 Median', min=0, max=100, value=0)
slider4 = widgets.IntSlider(description='H-Index - 3', min=0, max=100, value=0)
slider5 = widgets.IntSlider(description='G-Index - 3', min=0, max=100, value=0)
slider6 = widgets.IntSlider(description='Cites per year - 3', min=0, max=100, value=0)
slider7 = widgets.IntSlider(description='Cites per article - 3', min=0, max=100, value=0)
slider8 = widgets.IntSlider(description='Avg. Cites of top 10 papers', min=0, max=100, value=0)
slider9 = widgets.IntSlider(description='Avg. Altmetric Score of Top 10 Cited Papers', min=0, max=100, value=0)
slider10 = widgets.IntSlider(description='Avg. Altmetric Score of Top 10 Cited Papers', min=0, max=100, value=0)
slider11 = widgets.IntSlider(description='# of twitter followers', min=0, max=100, value=0)
slider12 = widgets.IntSlider(description='# of tweets', min=0, max=100, value=0)

interact(select_jour, w1=slider1, w2=slider2, w3=slider3, w4=slider4, w5=slider5, w6=slider6, w7=slider7, w8=slider8, \
         w9=slider9, w10=slider10, w11=slider11, w12=slider12)
```
